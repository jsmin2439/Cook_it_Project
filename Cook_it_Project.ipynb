```python
import pandas as pd
import matplotlib.pyplot as plt

# results.csv íŒŒì¼ ê²½ë¡œ (í•™ìŠµ í´ë”ì— ë”°ë¼ ê²½ë¡œê°€ ë‹¤ë¥¼ ìˆ˜ ìˆìŒ)
results_csv = "runs/detect/train/results.csv"

# CSV íŒŒì¼ ì½ê¸° (ì²« ë²ˆì§¸ í–‰ì€ í—¤ë”ì´ë¯€ë¡œ skiprows=0ì€ ìƒëµ)
df = pd.read_csv(results_csv)
print(df.tail(1))  # ë§ˆì§€ë§‰ ì—í¬í¬ì˜ ê²°ê³¼ í™•ì¸

# ì˜ˆì‹œ: 'train_loss', 'val_loss', 'mAP50' ì»¬ëŸ¼ì´ ìˆë‹¤ê³  ê°€ì •
plt.figure(figsize=(12, 5))

# train_lossë¥¼ ì„¸ ê°œì˜ ê°œë³„ ì†ì‹¤ ê°’ì˜ í•©ìœ¼ë¡œ ê³„ì‚°
df['train_loss'] = df['train/box_loss'] + df['train/cls_loss'] + df['train/dfl_loss']

# ë§Œì•½ validation ì†ì‹¤(val_loss)ë„ ê³„ì‚°í•˜ë ¤ë©´:
df['val_loss'] = df['val/box_loss'] + df['val/cls_loss'] + df['val/dfl_loss']


# í•™ìŠµ ë° ê²€ì¦ ì†ì‹¤ ê³¡ì„ 
plt.subplot(1, 2, 1)
plt.plot(df['epoch'], df['train_loss'], label='Train Loss')
plt.plot(df['epoch'], df['val_loss'], label='Validation Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs. Validation Loss")
plt.legend()

# mAP ê³¡ì„ 
plt.subplot(1, 2, 2)
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5', color='green')
plt.xlabel("Epoch")
plt.ylabel("mAP@0.5")
plt.title("Validation mAP@0.5")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from glob import glob

# í•™ìŠµ ë¡œê·¸ í´ë” ì„¤ì •
log_dir = "runs/detect/train"

# ê° ì§€í‘œì˜ ë¡œê·¸ íŒŒì¼ ê°€ì ¸ì˜¤ê¸°
loss_values = np.loadtxt(glob(f"{log_dir}/results.csv")[0], delimiter=",", skiprows=1, usecols=[2])  # ì†ì‹¤ê°’(Loss)
mAP50_values = np.loadtxt(glob(f"{log_dir}/results.csv")[0], delimiter=",", skiprows=1, usecols=[7])  # mAP50
```

```python
# ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(loss_values, label="Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(mAP50_values, label="mAP@0.5", color="green")
plt.xlabel("Epochs")
plt.ylabel("mAP@0.5")
plt.title("Mean Average Precision (mAP)")
plt.legend()

plt.show()
```
