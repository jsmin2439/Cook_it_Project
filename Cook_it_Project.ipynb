```python
import pandas as pd
import matplotlib.pyplot as plt

# results.csv 파일 경로 (학습 폴더에 따라 경로가 다를 수 있음)
results_csv = "runs/detect/train/results.csv"

# CSV 파일 읽기 (첫 번째 행은 헤더이므로 skiprows=0은 생략)
df = pd.read_csv(results_csv)
print(df.tail(1))  # 마지막 에포크의 결과 확인

# 예시: 'train_loss', 'val_loss', 'mAP50' 컬럼이 있다고 가정
plt.figure(figsize=(12, 5))

# train_loss를 세 개의 개별 손실 값의 합으로 계산
df['train_loss'] = df['train/box_loss'] + df['train/cls_loss'] + df['train/dfl_loss']

# 만약 validation 손실(val_loss)도 계산하려면:
df['val_loss'] = df['val/box_loss'] + df['val/cls_loss'] + df['val/dfl_loss']


# 학습 및 검증 손실 곡선
plt.subplot(1, 2, 1)
plt.plot(df['epoch'], df['train_loss'], label='Train Loss')
plt.plot(df['epoch'], df['val_loss'], label='Validation Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs. Validation Loss")
plt.legend()

# mAP 곡선
plt.subplot(1, 2, 2)
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5', color='green')
plt.xlabel("Epoch")
plt.ylabel("mAP@0.5")
plt.title("Validation mAP@0.5")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from glob import glob

# 학습 로그 폴더 설정
log_dir = "runs/detect/train"

# 각 지표의 로그 파일 가져오기
loss_values = np.loadtxt(glob(f"{log_dir}/results.csv")[0], delimiter=",", skiprows=1, usecols=[2])  # 손실값(Loss)
mAP50_values = np.loadtxt(glob(f"{log_dir}/results.csv")[0], delimiter=",", skiprows=1, usecols=[7])  # mAP50
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(loss_values, label="Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(mAP50_values, label="mAP@0.5", color="green")
plt.xlabel("Epochs")
plt.ylabel("mAP@0.5")
plt.title("Mean Average Precision (mAP)")
plt.legend()

plt.show()
```
